```python
import json
import time
import requests
from pathlib import Path

# rows = []
def opensearch_api(SEARCHNAME):
    S = requests.Session()
    URL = "https://en.wikipedia.org/w/api.php"
    if type(SEARCHNAME) == str and SEARCHNAME != '':
        SEARCHNAME = SEARCHNAME.capitalize()
    else:
        return
# Whether the keyword searched is type of string or not
    PARAMS = {
        "action": "opensearch",
        "namespace": "0",
        "search": SEARCHNAME,
        "limit": "3",
        "format": "json"
    }

    R = S.get(url=URL, params=PARAMS)
    DATA = R.json()
#     print(DATA)
    print(DATA[1])
    if DATA[1] != [] and DATA[1][0].capitalize() == SEARCHNAME:
        print(1)
# Whether found correct source from wiki api or not
        nameurl = DATA[3][0]
# The new url is the data from api and we will scrape people info from this web site. 
        try:
            for table in pd.read_html(nameurl):
                lis = table.iloc[:,0].values.tolist()
                if 'Born' in lis:
                    index = lis.index('Born')
#                     rows.append((SEARCHNAME, table.iloc[index, 1], nameurl))
#                     print(1, SEARCHNAME)
                    return (SEARCHNAME, table.iloc[index, 1], nameurl)
                elif 'Born:' in lis:
                    index = lis.index('Born:')
#                     rows.append((SEARCHNAME, table.iloc[index, 1], nameurl))
#                     print(1, SEARCHNAME)
                    return (SEARCHNAME, table.iloc[index, 1], nameurl)
                elif 'Based in' in lis:
                    index = lis.index('Based in')
#                     rows.append((SEARCHNAME, table.iloc[index, 1], nameurl))
#                     print(2, SEARCHNAME)
                    return (SEARCHNAME, table.iloc[index, 1], nameurl)
                elif 'Place of birth' in lis:
                    index = lis.index('Place of birth')
#                     rows.append((SEARCHNAME, table.iloc[index, 1], nameurl))
#                     print(3, SEARCHNAME)
                    return (SEARCHNAME, table.iloc[index, 1], nameurl)
                elif 'Headquarters' in lis:
                    index = lis.index('Headquarters')
                    rows.append((SEARCHNAME, table.iloc[index, 1], nameurl))
#                     print(4, SEARCHNAME)
                    return (SEARCHNAME, table.iloc[index, 1], nameurl)
                elif SEARCHNAME == 'Liza koshy':
                    rows.append((SEARCHNAME, 'Houston, Texas', nameurl))
#                     return (SEARCHNAME, 'Houston, Texas', nameurl)
            return None
# Grab kinds of desired data from web. 
#             print(SEARCHNAME)
        except (ValueError, ImportError, IndexError, AttributeError):
#             print(SEARCHNAME)
            return None

def get_borndf():
    filepath = 'ins.csv'
    if Path(filepath).exists():
        insdf = pd.read_csv(filepath)
    else: 
        insdf = get_insdf()
# Whether exist ins.csv to help as to build born.csv or not
    rows = []
    for row in insdf.itertuples():
        SEARCHNAME = row.name
        result = opensearch_api(SEARCHNAME)
        if result != None:
            rows.append(result)
# Pick name from ins.csv and then find info from wiki api 
    borndf = pd.DataFrame(rows, columns=["Name", "Born", "Website"])
#     borndf.to_csv("born.csv", mode='w',encoding='utf-8',header=True,index=False)
# Store data to csv file
    return borndf
# index 2

s = time.time()
borndf = get_borndf()
e = time.time()
e - s
```

```python
def get_us():
    url = 'https://worldpopulationreview.com/states'
# https://worldpopulationreview.com/us-cities
    content = requests.get(url)
    soup = BeautifulSoup(content.content, "html.parser")  
    res = list()
    for tr in soup.find('tbody').find_all('tr'):
        temp = list()
        td = tr.find_all('td')
        for j in range(9):
            if j == 1:
                temp.append(td[j].find('a').contents[0])
            elif j == 3 or j == 6:
                temp.append(td[j].find('span').contents[0])
            else:
                temp.append(td[j].contents[0])
        res.append(temp)
    usdf = pd.DataFrame(res)
    usdf.columns = ['Rank','State','2020 Pop','2020 Growth','2018 Pop','2010 Census','Growth Since 2010','% of US', 'Density (p/mi²)']
    return usdf
get_us()
```

```python
import os
import time
def store_data(df, filepath):
    partpath = filepath.split('/')[0]
    if not Path(partpath).exists():
        os.makedirs(partpath)
    df.to_csv(filepath, mode=""'w',encoding='utf-8',header=True,index=False)

def model_data():
    insdf = get_insdf()
    borndf = get_borndf()
    usdf = get_us()

    store_data(insdf, "data/ins.csv")
    store_data(borndf, "data/born.csv")
    store_data(usdf, "data/us_states.csv")
    return insdf, borndf, usdf
s = time.time()
insdf, borndf, usdf = model_data()
e = time.time()
e - s
```
